```python
import pandas as pd
from sklearn.ensemble import GradientBoostingClassifier
```

```python
backorder_train = pd.read_csv("D:\data ANALYTICS AND SCIENCE\PROJECT SUBMISSION DATA\manufacturing\product_train.csv")
backorder_train
```

```python
backorder_test = pd.read_csv("D:\data ANALYTICS AND SCIENCE\PROJECT SUBMISSION DATA\manufacturing\product_test.csv")
backorder_test
```

```python
#Define the feature columns and target variable
X_train = backorder_train.drop('went_on_backorder', axis=1)
y_train = backorder_train['went_on_backorder']
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder

# Define the column transformer
column_transformer = ColumnTransformer(
    transformers=[
        ('encoder', OneHotEncoder(), ['potential_issue', 'deck_risk', 'oe_constraint', 'ppap_risk', 'stop_auto_buy', 'rev_stop'])
    ])

dp_pipe = Pipeline([
    ('preprocessor', column_transformer)
])
```

```python
#Fit and transform the training data
train = dp_pipe.fit_transform(X_train)
```

```python
#Transform the test data
test = dp_pipe.transform(backorder_test)
```

```python
# Step 4: Create the gradient boosting classifier
gbm = GradientBoostingClassifier(random_state=42)
```

```python
from sklearn.model_selection import GridSearchCV

# Step 5: Define hyperparameter grid
param_grid = {
    'learning_rate': [0.1, 0.01, 0.001],
    'n_estimators': [50, 100, 200, 500, 700],
    'max_depth': [1, 2, 3, 4, 5, 6, 7],
    'min_samples_leaf': [1, 2, 5, 10]
}
```

```python
# Step 6: Create the grid search object
grid_search = GridSearchCV(estimator=gbm, param_grid=param_grid, cv=10)

# Step 7: Fit the grid search to the training data
grid_search.fit(train, y_train)

# Step 8: Get the best parameters
best_params = grid_search.best_params_
```

```python
best_params = {
'max_depth': 3,
'n_estimators': 700,
'learning_rate': 0.1,
'min_samples_leaf': 20,
'subsample': 0.9
}
```

```python
gbm_model = GradientBoostingClassifier(**best_params)
```

```python
#Fit the model on the training data
gbm_model.fit(train, y_train)
```

```python
# Step 4: Fit the gradient boosting classifier
gbm.fit(train, y_train)

# Step 5: Make predictions on the test data
test_score = gbm.predict_proba(test)[:, 1]
test_score

```

```python
import scipy.stats as stats

# Calculate the mode of test_score
mode_score = stats.mode(test_score, keepdims=True)

print(mode_score.mode[0])

```

```python
# Define the threshold
threshold = 0.0066

# Convert probabilities to "Yes" or "No" predictions
yes_no = ['Yes' if score >= threshold else 'No' for score in test_score]

print(yes_no)


```
